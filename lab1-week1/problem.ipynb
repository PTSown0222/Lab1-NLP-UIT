### Bài tập 3: Biểu thức chính quy

Viết một hàm có tên `find_and_replace_dates` để tìm tất cả các ngày có định dạng `mm/dd/yyyy` trong một chuỗi văn bản đầu vào và thay thế chúng bằng định dạng `yyyy-mm-dd`. Hàm sẽ trả về một `tuple` bao gồm danh sách (`list`) các ngày tìm được và chuỗi (`str`) văn bản mới sau khi thực hiện thay thế.

```python
### BEGIN SOLUTION
from typing import List, Tuple
import re

def find_and_replace_dates(corpus: str) -> Tuple[List[str], str]:
    # Tìm tất cả ngày theo pattern mm/dd/yyyy
    dates = re.findall(r'(\d+)/(\d+)/(\d+)', corpus)
    
    # onlyDate: giữ format mm/dd/yyyy
    onlyDate = [f"{m}/{d}/{y}" for m, d, y in dates]
    
    # corpus_fixed: chuyển mm/dd/yyyy -> yyyy-mm-dd
    corpus_fixed = re.sub(
        r'(\d+)/(\d+)/(\d+)',
        r'\3-\1-\2',  # \3 = year, \1 = month, \2 = day
        corpus
    )
    
    return onlyDate, corpus_fixed

### END SOLUTION
```

Hàm sẽ sắp xếp các mã chứng khoán theo thứ tự tăng dần và sau đó nối chúng lại thành một chuỗi mới, với các mã vẫn được phân tách bởi dấu phẩy. Hàm sẽ trả về một chuỗi (`str`).

```python
### BEGIN SOLUTION
def sort_and_join_stocks(text : str) -> str:
    return ','.join(sorted(text.split(',')))
### END SOLUTION
```

### Bài tập 7: Làm việc với tập tin

Các cột trong tệp `./data/portfolio.csv` tương ứng với `tên cổ phiếu`, `số lượng cổ phiếu` và `giá mua của một lô cổ phiếu đơn lẻ`. Đọc tất cả các dòng và tính toán xem việc mua tất cả các cổ phiếu trong danh mục đầu tư này tốn bao nhiêu tiền và lưu kết quả vào biến `total_cost`. Gợi ý: để chuyển đổi một chuỗi thành một số nguyên, sử dụng `int(s)`. Để chuyển đổi một chuỗi thành một số thực, sử dụng `float(s)`. Nếu bạn sử dụng thư viện `pandas` trong bài tập này, bạn sẽ không được tính điểm.

```python
total_cost = None

### BEGIN SOLUTION
def find_path(name : str, start_dirs= None) -> str:

    # Kiểm tra nếu không có đường dẫn thì trả về đường dẫn hiện tại
    if start_dirs is None:
        start_dirs = os.getcwd()

    # Kiểm tra trong thư mục gốc

    # ghép đường dẫn hiện tại + tên file/folder
    current_working_directory = os.path.join(start_dirs, name)
    # Kiểm tra xem có tồn tại đường dẫn hiện tại + files/folder
    
    # Nếu có thì trả về đúng đường dẫn 
    if os.path.exists(current_working_directory):
        return current_working_directory
    

    # Tìm trong các thư mục con
    for root, dirs, files in os.walk(start_dirs):
        
        # Kiểm tra trong danh sách file
        if name in files:
            return os.path.join(root, name)
        
        # Kiểm tra trong danh sách folder
        if name in dirs:
            return os.path.join(root, name)

    return None

def read_portfolio(file_name):
    # Tìm đường dẫn
    path = find_path(file_name)
    total = 0
    # Đọc và tính tổng
    with open(path, 'r', encoding='utf-8') as f:
        # Bỏ header
        next(f)
        # Đọc từng dòng
        for line in f:
            parts = line.strip().split(',')
            if len(parts) >= 3:
                shares = int(parts[1])
                price = float(parts[2])
                total += shares * price
    
    return total

total_cost = read_portfolio('./data/portfolio.csv')
print(int(total_cost))
### END SOLUTION
```

**Yêu cầu:**
1.  **Tạo cột mới:** Thêm một cột có tên là `market_value` vào `portfolio_df`. Giá trị của mỗi hàng trong cột này được tính bằng cách nhân giá trị của cột `shares` với cột `price`.
